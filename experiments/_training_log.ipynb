```python
import json
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
model_name = ['m004_20230215_152506', 'm006_20230216_104829', 'm008_20230216_134648', 'm009_20230220_150313']
logs_json_dir = [Path('logs_json', n) for n in model_name]
```

```python
def showTrainingLog(_dir, epochs=500, st_epoch=1):
    for i, key in enumerate(['train_loss', 'test_loss', 'train_acc', 'test_acc']):
        print(key)
        plt.figure(figsize=(5, 5))
        
        for d in _dir:
            with open(d/f'epoch_{st_epoch}-{epochs}.json') as f:
                logs = json.load(f)

            plt.plot(range(st_epoch, st_epoch+epochs), logs[key])
            plt.xlabel('Epoch')
            if i in [0, 1]:
                plt.ylabel('Loss')
                plt.ylim(0.2, 1.3)
            else:
                plt.ylabel('Accuracy')
                plt.ylim(0.62, 0.93)
        if i in [0, 1]:
            loc = 'upper right'
        else:
            loc = 'lower right'
        
        plt.xlim(st_epoch, epochs)
        plt.legend([f'Model {m}' for m in range(1, 5)], loc=loc)
        plt.tight_layout()
        plt.show()
```

```python
showTrainingLog(logs_json_dir)
```
